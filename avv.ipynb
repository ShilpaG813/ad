```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
df = pd.read_csv(r'C:\Users\shilp\Onedrive\Desktop\titan.csv')
df.isna().sum()
df=df.dropna()
dp=df['Age'].fillna(df['Age'].mean())
df.isna().sum()

features = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
ft = df['Fare']
survival_target = df['Survived']

features = pd.get_dummies(features, columns=['Sex'], drop_first=True)
scaler = StandardScaler()
features = scaler.fit_transform(features)

X_train, X_test, fare_train, fare_test, survival_train, survival_test = train_test_split(
    features, ft, survival_target, test_size=0.2, random_state=42
)

input_layer = keras.layers.Input(shape=(features.shape[1],))
hidden_layer = keras.layers.Dense(64, activation='relu')(input_layer)
hidden_layer = keras.layers.Dense(32, activation='relu')(hidden_layer)

# Regression output
fare_output = keras.layers.Dense(1, name='regression')(hidden_layer)

# Classification output
survival_output = keras.layers.Dense(1, activation='sigmoid', name='classification')(hidden_layer)

model = keras.models.Model(inputs=input_layer, outputs=[fare_output, survival_output])

# Compile the model with different loss functions for regression and classification
model.compile(
    optimizer='adam',
    loss={'regression': 'mean_squared_error', 'classification': 'binary_crossentropy'},
    metrics={'classification': 'accuracy'},
)

model.fit(
    X_train,
    {'regression': fare_train, 'classification': survival_train},
    epochs=10,
    batch_size=32,
    validation_split=0.2,
)

results = model.evaluate(X_test, {'regression': fare_test, 'classification': survival_test})
fare_loss, survival_loss = results[0], results[2]
survival_accuracy = results[3]

print(f"\nFare Model Loss: {fare_loss}")
print(f"\nSurvival Model Loss: {survival_loss}, Survival Model Accuracy: {survival_accuracy}")
```
